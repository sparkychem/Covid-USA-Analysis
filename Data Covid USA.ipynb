```python
import pandas as pd
```

```python
df = pd.read_csv('USA Covid Data.csv')
df.head()
```

```python
# persentase total kematian terhadap total kasus
df['Percentage Deaths per Total Cases'] = df['Total Deaths'] * 100 / df['Total Cases']
df.sort_values(['Percentage Deaths per Total Cases'], ascending = False)
```

```python
# Pennsylvania merupakan daerah dengan presentasi kematian tertinggi, sedangkan alaska yang terendah
```

```python
# persentase total kesembuhan terhadap total kasus
df['Percentage Recovered per Total Cases'] = df['Total Recovered'] * 100 / df['Total Cases']
df.sort_values(['Percentage Recovered per Total Cases'], ascending = False)
```

```python
# Minnesota merupakan daerah dengan tingkat kesembuhan tertinggi, dan indiana terendah
```

```python
# persentase kasus aktif terhadap total populasi
df['Percentage Active Cases per Population'] = df['Active Cases'] * 100 / df['Population']
df.sort_values(['Percentage Active Cases per Population'], ascending = False)
```

```python
# Indiana merupakan daerah dengan persentasi kasus aktif tertinggi dan minnesota terendah
```

```python
# Skor penangan Covid di AS
df['Score'] = df['Percentage Recovered per Total Cases'] - (df['Percentage Deaths per Total Cases'] + df['Percentage Active Cases per Population'])
df.sort_values(['Score'], ascending = False)
```

```python
# 3 daerah dengan penanganan Covid terbaik: Minnesota, New Hampshire dan North Dakota
# 3 daerah dengan penanganan Covid terburuk: Indiana, New Jersey dan Oregon
```

```python
# Apakah ada korelasi antara jumlah tes dengan skor penangan covid?
df.corr()
```
